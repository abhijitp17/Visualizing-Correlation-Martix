# Visualizing a correlation matrix in Python

A correlation matrix is a table showing correlation coefficients between sets of variables. Each random variable (Xi) in the table is correlated with each of the other values in the table (Xj).

```python
#Reading packages in library
import pandas as pd
import numpy as np
```

```python
#Creating random dataset
#rds = Random Dataset
rds = np.random.RandomState(0)
```

```python
#Creating dataframe for random dataset
df = pd.DataFrame(rds.rand(10,10))
df.head()
```

```python
#Creating correlation matrix
corr = df.corr()
print(corr)
```

```python
#Correlation matrix visualization
corr.style.background_gradient(cmap = 'coolwarm')
```

```python
# Limiting decimal number to 2
corr.style.background_gradient(cmap = 'coolwarm').set_precision(2)
```

```python
#Highlighting highest number in column
df.style.highlight_max(axis=0)
```

```python
#Highlighting highest number in row
df.style.highlight_max(axis=1)
```

```python
#Color negative numbers red
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color

# Hightlight large numbers
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

# Create correlation matrix:
rds = np.random.RandomState(0)
df = pd.DataFrame(rs.rand(10, 10))
corr = df.corr()
corr.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```
